```python
# question 5

# Here is the ChatBot's explanation/summary of the difference between
# attributes and methods:
# Attributes: Think of attributes as pieces of information or properties that
# belong to an object. They give you details about the object’s state but don’t
# do anything beyond that. For instance, df.shape tells you the dimensions of 
# the DataFrame without changing anything.
# Methods: Methods are more like actions or operations that an object can perform. 
# They do something with the object’s data and often return a result. For example, 
# df.describe() performs a calculation and provides a summary of statistics about 
# the DataFrame's numerical columns.

# My takeaway: 
# Attributes are simply values that are associated with the object. They hold data 
# about the object and when called may give information about the object itself.
# Methods on the otherhand are functions associated with the object and can 
# perform actions based on the object's data or even modify it's state.
```

```python
# question 6

#definitions:
# Count: Tells the number of valid data entries in the column
# Mean: Sum of all values divided by the count of non-missing values
# Std: Std stands for Standard Deviation and this measures the amount of 
# dispersion of the values from the mean, helping us to understand if the 
# data is consistent
# Min: The minimum/smallest value in the column
# 25%: The 25th percentile divides the lowest 25% of data from the rest by 
# giving the value where the data seperates
# 50%: The 50th percentile divides the data into two equal halves and gives 
# value at the middle of the data when sorted in ASCENDING order
# 75%: The 75th percentile seperates the lowest 75% of data from the highest 
# 25% by giving the value at the position where the data seperates
# Max: The maximum/largest value in the column
```

```python
# question 7

# df.dropna() removes the rows missing values by default
# but can be altered to also remove columns with missing vaues by using 
# the (axis=1) parameter.
# del df['col'] deletes specific columns from the dataframe regardless of 
# whether or not it is missing values.
# Both are useful, but df.dropna() is useful if you're trying to get rid of
# all rows or columns with missing values while del df['col'] is useful if 
# you want to get rid of a specific column (possibly containing irrelevant 
# data.

# 1.
# A use case where df.dropna() might be preferred over del df['col'] when you 
# require full/complete sets of data. Removing the rows or columns with incomplete
# info will allow you to keep your dataframe consistent.

# 2.
# A use case where del df['col'] might be more useful may be when a specific column 
# is not useful/irrelevant. del df['col'] will simplify your dataframe.

# 3. 
# First using del df['col'] could be important because removing irrelevant columns first
# makes the data cleaning process cleaner since working with only important columns streamlines 
# the focus on cleaning.

# 4. 
import pandas as pd

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv"
df = pd.read_csv(url)
df.isnull().sum()

# In this dataset example, by using df.isnull().sum(), I can see that there is a 
# significant amount of data missing in the 'deck' column. I potentially may want 
# to remove all rows with missing data and the 'deck' column ENTIRELY because since 
# there is so much missing info, the data in that column is not consistent. The logical 
# thing to do, would be to use del df['deck'] first.
```
